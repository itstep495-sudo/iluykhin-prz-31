# Система выделения и маскирования объектов на изображении
Алгоритм работы:

*   Пользователь загружает изображение
*   OpenCV выполняет обработку изображения
*   С помощью детектора Haar Cascade находятся лица
*   Обнаруженные области автоматически закрашиваются
*   Результат выводится пользователю

Практическое применение:

* скрытие персональных данных
* анонимизация лиц
* модерация изображений
* предварительная обработка данных для ML-систем.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

image_name = list(uploaded.keys())[0]

file_bytes = np.asarray(
    bytearray(uploaded[image_name]),
    dtype=np.uint8
)

img = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)
```

```python
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades +
    'haarcascade_frontalface_default.xml'
)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5
)
```

```python
for (x, y, w, h) in faces:

    cv2.rectangle(
        img,
        (x, y),
        (x+w, y+h),
        (0, 0, 0),
        -1
    )
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,8))
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
